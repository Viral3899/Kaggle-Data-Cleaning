### Football Dataset 

```python
#Importing Dependencies
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading Dataset
df=pd.read_csv('NFL Play by Play 2009-2016 (v3).csv')
df.head()
```

```python
#Finding Null Values for Each Column
missing_values_count = df.isnull().sum()

# look at the # of missing points in the first ten columns
missing_values_count[0:10]
```

### Percent of Missing Value

```python
#Finding Total Cell Counts
total_cells=np.product(df.shape)

#Finding Total Count of Null
total_missing=missing_values_count.sum()

# Percent Missing
Percent_Missing=(total_missing/total_cells)*100
Percent_Missing
```

```python
# Droping Rows Having Null Values
df.dropna()
```

#### All Rows have Null Values so no data Left

```python
# Droping Columns Having Null Values
columns_with_na_dropped=df.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
print(f"total no of Columns {len(df.columns)}")
print(f"no. of Columns After droping {len(columns_with_na_dropped.columns)}")
```

### There are Large number Columns Dropped 

### Making Subset of Data

```python
subset_df = df.loc[:, 'EPA':'Season']
subset_df.head()
```

```python
# Filling Null With 0

subset_df.fillna(0)
```

```python
# replace all Null's the value that comes directly after it in the same column, & Others with 0

subset_df.fillna(method='bfill', axis=0).fillna(0)
```
